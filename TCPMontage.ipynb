```python
import mne, os
```

```python
dirpath = '/home/krakar/Files/uni/tukl/data/TUH EEG Corpus/edf/eval/normal/01_tcp_ar'
```

```python
file = os.path.join(dirpath, os.listdir(dirpath)[0])
```

```python
# Load the EEG data from an EDF file
edf_file = file
raw = mne.io.read_raw_edf(edf_file, preload=True)

# Define the electrode positions for the TCP montage
montage = mne.channels.make_standard_montage('standard_1020')

# Apply the montage to the raw data
raw.set_montage(montage)

# Visualize the montage to ensure it's set up correctly (optional)
raw.plot_sensors(kind='topomap', show_names=True)

# Save the data with the new montage (optional)
raw.save('path_to_filtered_tcp_montage_data.fif', overwrite=True)
```

```python
pair = ('EEG FP1-REF', 'EEG F7-REF')
anode = pair[0]
cathode = pair[1]
```

```python
raw2 = mne.set_bipolar_reference(raw.load_data(), anode=[pair[0]], cathode=[pair[1]], ch_name=f'{anode}-{cathode}', drop_refs=False, copy=True, verbose=False)
```
