```python
from datasets.dataset import Dataset
from datasets.pytordataset import EEGDataset
from datasets.pipeline import Pipeline, ClipData, CropData, ReduceChannels, ResampleData
```

```python
datapipeline = Pipeline()
datapipeline.add(ClipData(100))
datapipeline.add(CropData(60, 480))
datapipeline.add(ReduceChannels())
datapipeline.add(ResampleData(100))
```

```python
tuhdata = Dataset('/home/krakar/Files/uni/tukl/data')
tuhdata.traindir
```

```python
tuhdata.set_pipeline(datapipeline)
```

```python
dest_dir = "data"
tuhdata.save_to_npz(dest_dir, 'eval')
```

```python
checkdata = EEGDataset('data/eval')
from torch.utils.data import DataLoader
```

```python
eval_dataloader = DataLoader(checkdata)
```

```python
train_features, train_labels = next(iter(eval_dataloader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
```
